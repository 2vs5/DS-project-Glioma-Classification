```python
import pickle
import pandas as pd
from pandas import read_csv
import numpy as np
from scipy.stats import chi2_contingency, fisher_exact
```

```python
# Patient IDs
GBM = pickle.load(open('/content/drive/MyDrive/DS_project/Data/5_Inter-cohort_preprocessing/5.1.data_spliting/GBM_data_split_result.pickle', 'rb'))
LGG = pickle.load(open('/content/drive/MyDrive/DS_project/Data/5_Inter-cohort_preprocessing/5.1.data_spliting/LGG_data_split_result.pickle', 'rb'))

# Patient metadata
GBM_meta = read_csv("/content/drive/MyDrive/DS_project/Data/0.clinical_rdata_to_csv/TCGA-GBM_clinicalXML_patient.csv")
LGG_meta = read_csv("/content/drive/MyDrive/DS_project/Data/0.clinical_rdata_to_csv/TCGA-LGG_clinicalXML_patient.csv")
```

```python
# Load the patient IDs
GBM_train_pid = [sample[:-4] for sample in GBM["Train"]]
LGG_train_pid = [sample[:-4] for sample in LGG["Train"]]

# Load the gender information
GBM_gender = GBM_meta[GBM_meta["bcr_patient_barcode"].isin(GBM_train_pid)][["bcr_patient_barcode","gender"]]
LGG_gender = LGG_meta[LGG_meta["bcr_patient_barcode"].isin(LGG_train_pid)][["bcr_patient_barcode","gender"]]
```

```python
# Count the number of male and female
GBM_male, GBM_female = GBM_gender["gender"].value_counts()
LGG_male, LGG_female = LGG_gender["gender"].value_counts()
```

```python
# 2x2 contingency table
# Exclude samples with unknown gender (n=4)
contingency_table = np.array([
    [LGG_male, GBM_male],  # Male: LGG, GBM
    [LGG_female, GBM_female]   # Female: LGG, GBM
])
```

```python
# Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-square Test")
print("Test statistic:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)
```

```python
# Fisher's exact test
oddsratio, p_fisher = fisher_exact(contingency_table)
print("\nFisher's Exact Test")
print("Odds ratio:", oddsratio)
print("p-value:", p_fisher)
```
